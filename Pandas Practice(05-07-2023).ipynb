# .To select any specific rows with index postion

# loc    AND    iloc   in pandas 


The loc method allows you to access rows and columns in a DataFrame using labels.
It uses the syntax df.loc[row_labels, column_labels].
You can specify single labels, lists of labels, or even slices (start_label:end_label).
It is inclusive of both the start and end labels.

```python
# Accessing Single row from data frame
df.loc[0]
```

```python
df.loc[1]
```

```python
#accessing multiple rows from data frame
df.loc[2:3]
```

```python
#Accessing specific columns using labels in this example we get row number 1 to 3 and 2 column

df.loc[1:3,["Student_Name",'Student_Reg_Num']]
```
